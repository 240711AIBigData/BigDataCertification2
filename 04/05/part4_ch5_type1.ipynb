```python
import pandas as pd
df = pd.read_csv('./data/data5-1.csv')
print(df.head())
```

### âœï¸ bmi ìˆ˜ì¹˜ê°€ ì •ìƒì¸ ì‚¬ëžŒ ìˆ˜ì™€ ìœ„í—˜ì²´ì¤‘ì¸ ì‚¬ëžŒ ìˆ˜ì˜ ì°¨ì´ë¥¼ ì ˆëŒ“ê°’ìœ¼ë¡œ êµ¬í•˜ì‹œì˜¤. (ì •ìˆ˜ë¡œ ì¶œë ¥)
- bmi(ì²´ì§ˆëŸ‰ ì§€ìˆ˜) : ëª¸ë¬´ê²Œ(kg) / í‚¤(m) ì˜ ì œê³± ë‹¨ìœ„
- ë‹¨ìœ„ : Height(cm), Weight(kg)
- BMI
    - ì €ì²´ì¤‘ : BMI 18.5 ë¯¸ë§Œ
    - ì •ìƒì²´ì¤‘ : BMI 18.5 ì´ìƒ 23 ë¯¸ë§Œ
    - ê³¼ì²´ì¤‘ ë˜ëŠ” ìœ„í—˜ ì²´ì¤‘ : BMI 23 ì´ìƒ 25 ë¯¸ë§Œ
    - ë¹„ë§Œì²´ì¤‘ : 25 ì´ìƒ

```python
import pandas as pd
df = pd.read_csv('./data/data5-2.csv')
print(df.head())
```

```python
# 1. bmi íŒŒìƒë³€ìˆ˜ ê³„ì‚°
## íŒŒìƒë³€ìˆ˜ë¥¼ ê³„ì‚°í•˜ê¸° ì•žì„œ ë‹¨ìœ„ì˜ ë³€í™˜ í•„ìš” : ë¬¸ì œì—ì„œ bmi ê³„ì‚° ìˆ˜ì‹ì„ ë³´ë©´ í‚¤ëŠ” m ë‹¨ìœ„
## í˜„ìž¬ Height ëŠ” ë‹¨ìœ„ê°€ cm ë¼ì„œ m ë‹¨ìœ„ë¡œ ë³€ê²½í•œ í›„, ì œê³±ì—°ì‚°ìž(**) ì‚¬ìš©
df['bmi'] = df['Weight'] / (df['Height'] / 100)**2
print(df.head(), '\n')

# 2. ì¡°ê±´1 : ì •ìƒì²´ì¤‘ bmi
cond1 = (df['bmi'] >= 18.5) & (df['bmi'] < 23)

# 3. ì¡°ê±´2 : ìœ„í—˜ì²´ì¤‘ bmi
cond2 = (df['bmi'] >= 23) & (df['bmi'] < 25)

# 4. ì¡°ê±´1ì˜ ì¸ì›ê³¼ ì¡°ê±´2ì˜ ì¸ì› ì°¨ì´ ê³„ì‚°, ì ˆëŒ“ê°’ ì²˜ë¦¬
## ì •ìƒì²´ì¤‘ ì¸ì›ê³¼ ìœ„í—˜ì²´ì¤‘ ì¸ì›ì˜ ì°¨ì´ë¥¼ ê³„ì‚°í•˜ê³  ì ˆëŒ“ê°’ìœ¼ë¡œ ì¶œë ¥
print(abs(len(df[cond1]) - len(df[cond2])))
```
