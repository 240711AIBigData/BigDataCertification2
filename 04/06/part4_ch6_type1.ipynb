### âœï¸ ì¶œë™ì‹œê°„ê³¼ ë„ì°©ì‹œê°„ ì°¨ì´ê°€ í‰ê· ì ìœ¼ë¡œ ê°€ì¥ ì˜¤ë˜ ê±¸ë¦° ì†Œë°©ì„œì˜ ì‹œê°„ì„ ë¶„ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ ì¶œë ¥í•˜ì‹œì˜¤.

```python
import pandas as pd
df = pd.read_csv('./data/data6-1-1.csv')
print(df.head())
```

```python
# 1. datetime í˜•íƒœë¡œ ë³€ê²½
## ë„ì°©ì‹œê°„ê³¼ ì¶œë™ì‹œê°„ì„ object ì—ì„œ datetime ìë£Œí˜•ìœ¼ë¡œ ë³€ê²½
df['ì¶œë™ì‹œê°„'] = pd.to_datetime(df['ì¶œë™ì‹œê°„'])
df['ë„ì°©ì‹œê°„'] = pd.to_datetime(df['ë„ì°©ì‹œê°„'])

# 2. ì¶œë™ì‹œê°„ê³¼ ë„ì°©ì‹œê°„ ì°¨ì´ë¥¼ ë¶„ìœ¼ë¡œ ê³„ì‚°
## dt.total_seconds() í•¨ìˆ˜ëŠ” ì „ì²´ ì‹œê°„ì„ ì´ˆ ë‹¨ìœ„ë¡œ ë³€í™˜ -> 60ìœ¼ë¡œ ë‚˜ëˆ„ë©´ ë¶„
df['ì‹œê°„ì°¨ì´(ë¶„)'] = (df['ë„ì°©ì‹œê°„'] - df['ì¶œë™ì‹œê°„']).dt.total_seconds() / 60
print(df.head(), '\n')

# 3. ì†Œë°©ì„œë³„ í‰ê·  ì‹œê°„ ì°¨ì´ ê³„ì‚°
## ì†Œë°©ì„œë³„ ì‹œê°„ ì°¨ì´ì˜ í‰ê·  êµ¬í•˜ê¸°
## groupby('ì†Œë°©ì„œ') ì—ì„œëŠ” ì†Œê´„í˜¸ë¥¼ ì‚¬ìš©í•´ ê·¸ë£¹í•‘ ê¸°ì¤€ ì»¬ëŸ¼ ì„ íƒ
## ['ì‹œê°„ì°¨ì´(ë¶„)'].mean() í™œìš©í•´ ì‹œê°„ ì°¨ì´(ë¶„) ì»¬ëŸ¼ì˜ í‰ê· ë§Œ êµ¬í•˜ê¸°
avg_diff = df.groupby('ì†Œë°©ì„œ')['ì‹œê°„ì°¨ì´(ë¶„)'].mean()
print(avg_diff, '\n')

# 4. í‰ê·  ì°¨ì´ê°€ ê°€ì¥ í° ì†Œë°©ì„œì˜ ì‹œê°„ì„ ì°¾ê³  ì¶œë ¥
## idxmax() ëŠ” ê°€ì¥ í° ê°’ì˜ ì¸ë±ìŠ¤ëª…(ê°’)ì„ ë°›í™˜ë°›ìŒ
## 'ì†Œë°©ì„œ9' ê°’ì„ ê°–ê³  ìˆëŠ” ì¸ë±ìŠ¤ì˜ ê°’ ì¶œë ¥
## ì¶œë ¥ ê²°ê³¼ë¥¼ round() ë¡œ ë°˜ì˜¬ë¦¼í•´ ì •ìˆ˜ ì¶œë ¥
idx = avg_diff.idxmax()
result = avg_diff[idx]
print(result)
print(round(result))
```

### âœï¸ í•™êµì—ì„œ êµì‚¬ í•œ ëª…ë‹¹ ë§¡ì€ í•™ìƒ ìˆ˜ê°€ ê°€ì¥ ë§ì€ í•™êµë¥¼ ì°¾ê³ , ê·¸ í•™êµì˜ ì „ì²´ êµì‚¬ ìˆ˜ë¥¼ êµ¬í•˜ì‹œì˜¤.

```python
import pandas as pd
df = pd.read_csv('./data/data6-1-2.csv')
print(df.head())
```

```python
# [í’€ì´1]
# 1. ì´ í•™ìƒ ìˆ˜ ê³„ì‚°
## 1í•™ë…„ë¶€í„° 6í•™ë…„ê¹Œì§€ ë”í•œ ê°’ì„ ì´ í•™ìƒ ìˆ˜ë¡œ ê³„ì‚°
df['ì´í•™ìƒìˆ˜'] = df.iloc[:, 2:].sum(axis=1)
print(df, '\n')

# 2. êµì‚¬ í•œ ëª…ë‹¨ ë§¡ì€ í•™ìƒ ìˆ˜
## ì´ í•™ìƒ ìˆ˜ / êµì‚¬ ìˆ˜ë¡œ êµì‚¬ í•œ ëª…ë‹¹ ë§¡ì€ í•™ìƒì˜ ìˆ˜ ê³„ì‚°
df['í•™ìƒ/êµì‚¬'] = df['ì´í•™ìƒìˆ˜'] / df['êµì‚¬ìˆ˜']
print(df.head(), '\n')

# 3. í•™ìƒ/êµì‚¬ ì»¬ëŸ¼ì„ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬
## 'í•™ìƒ/êµì‚¬' ì»¬ëŸ¼ì„ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ë©´ 'í•™ìƒ/êµì‚¬' ì»¬ëŸ¼ì—ì„œ ê°€ì¥ í° ê°’ì´ ì²« ë²ˆì§¸ í–‰
df = df.sort_values('í•™ìƒ/êµì‚¬', ascending=False)
print(df.head(), '\n')

# 4. ìµœìƒë‹¨ í–‰ì˜ êµì‚¬ ìˆ˜ ê°’ ì¶œë ¥
## ì •ë ¬ì„ í•˜ê³  ë‚˜ë©´ ì¸ë±ìŠ¤ëª…(ê°’)ì´ ë’¤ì„ì´ê²Œ ë˜ë¯€ë¡œ iloc ì‚¬ìš©
## ì²«ë²ˆì§¸ í–‰ì˜ êµì‚¬ ìˆ˜ ê°’ì„ ì¶œë ¥
print(df.iloc[0, 1])
```

```python
import pandas as pd
df = pd.read_csv('./data/data6-1-2.csv')

# [í’€ì´2]
# 1. ì´ í•™ìƒ ìˆ˜ ê³„ì‚°
## 1í•™ë…„ë¶€í„° 6í•™ë…„ê¹Œì§€ ë”í•œ ê°’ì„ ì´ í•™ìƒ ìˆ˜ë¡œ ê³„ì‚°
df['ì´í•™ìƒìˆ˜'] = df.iloc[:, 2:].sum(axis=1)
print(df, '\n')

# 2. êµì‚¬ í•œ ëª…ë‹¨ ë§¡ì€ í•™ìƒ ìˆ˜
## ì´ í•™ìƒ ìˆ˜ / êµì‚¬ ìˆ˜ë¡œ êµì‚¬ í•œ ëª…ë‹¹ ë§¡ì€ í•™ìƒì˜ ìˆ˜ ê³„ì‚°
df['í•™ìƒ/êµì‚¬'] = df['ì´í•™ìƒìˆ˜'] / df['êµì‚¬ìˆ˜']
print(df.head(), '\n')

# 3. í•™ìƒ/êµì‚¬ ìµœëŒ€ê°’ì˜ ì¸ë±ìŠ¤ëª…
## 2ë²ˆì˜ ìµœëŒ“ê°’ì„ ê°€ì§„ í–‰ì„ idxmax() ë¡œ êµ¬í•œë‹¤
## idxmax() í™œìš©ì‹œ ì£¼ì˜ì  : 'ì¸ë±ìŠ¤ëª…'ì„ ë°˜í™˜í•œë‹¤
idx = df['í•™ìƒ/êµì‚¬'].idxmax()

# 4. í•™ìƒ/êµì‚¬ ìµœëŒ€ê°’ì˜ í•™êµ êµì‚¬ ìˆ˜ ì¶œë ¥
## ì¸ë±ìŠ¤ëª…ìœ¼ë¡œ ê°’ì„ ì°¾ê¸° ìœ„í•´ì„œëŠ” loc í™œìš©
## êµì‚¬ ìˆ˜ë¥¼ ë¬»ê³  ìˆìœ¼ë¯€ë¡œ idxmax() ë¡œ ë°˜í™˜ë°›ì€ ì¸ë±ìŠ¤ëª…ê³¼ 'êµì‚¬ ìˆ˜' ì»¬ëŸ¼ì„ ì„ íƒí•´ ê°’ ì¶œë ¥
print(df.loc[idx, 'êµì‚¬ìˆ˜'])
```

### âœï¸ ì—°ë„ë³„ë¡œ ì´ ë²”ì£„ ê±´ìˆ˜(ë²”ì£„ ìœ í˜•ë³„ ë°œìƒ ê±´ìˆ˜ì˜ í•©)ì˜ ì›”í‰ê·  ê°’ì„ êµ¬í•œ í›„ ê·¸ ê°’ì´ ê°€ì¥ í° ì—°ë„ë¥¼ ì°¾ì•„ í•´ë‹¹ ì—°ë„ì˜ ì´ ë²”ì£„ ê±´ìˆ˜ì˜ ì›”í‰ê·  ê°’ì„ ì¶œë ¥í•˜ì‹œì˜¤.(ë°˜ì˜¬ë¦¼í•˜ì—¬ ì •ìˆ˜ ì¶œë ¥)

```python
import pandas as pd
df = pd.read_csv('./data/data6-1-3.csv')
print(df.head())
```

```python
# [í’€ì´1]
# 1. ì´ ë²”ì£„ ê±´ìˆ˜ ê³„ì‚°
## ë²”ì£„ ìœ í˜•ë³„ ë°œìƒ ê±´ìˆ˜ë¥¼ í•©íˆê³ , ìƒˆë¡œìš´ ì»¬ëŸ¼ì— ëŒ€ì… -> í–‰ë³„ í•©ì„ ìœ„í•´ ì¶•(axis)ì„ 1ë¡œ ì„¤ì •
df['ì´ë²”ì£„ê±´ìˆ˜'] = df.iloc[:, 1:7].sum(axis=1)
print(df.head(), '\n')

# 2. ì—°ë„ ìŠ¬ë¼ì´ì‹±
df['ì—°ë„'] = df['ë‚ ì§œ'].str[:4]
print(df.head(), '\n')

# 3. ì—°ë„ë³„ ì´ ë²”ì£„ ê±´ìˆ˜ í•© ê³„ì‚°
## ì´ ë²”ì£„ ê±´ìˆ˜ë¥¼ ì—°ë„ë³„ë¡œ ê·¸ë£¹í•‘í•˜ê³  í•©í•˜ê¸°
result = df['ì´ë²”ì£„ê±´ìˆ˜'].groupby(df['ì—°ë„']).sum()
print(result, '\n')

# 4. ê°€ì¥ í° ê°’ì˜ ì›”í‰ê·  ê³„ì‚°
## result ì—ì„œ ê°€ì¥ í° ê°’ì„ 12ë¡œ ë‚˜ëˆ„ê³  ë°˜ì˜¬ë¦¼ í›„ ì¶œë ¥
print(round(result.max()/12))
```

```python
# [í’€ì´2]
# 1. ì´ ë²”ì£„ ê±´ìˆ˜ ê³„ì‚°
## ë²”ì£„ ìœ í˜•ë³„ ë°œìƒ ê±´ìˆ˜ë¥¼ í•©íˆê³ , ìƒˆë¡œìš´ ì»¬ëŸ¼ì— ëŒ€ì… -> í–‰ë³„ í•©ì„ ìœ„í•´ ì¶•(axis)ì„ 1ë¡œ ì„¤ì •
df['ì´ë²”ì£„ê±´ìˆ˜'] = df.iloc[:, 1:7].sum(axis=1)
print(df.head(), '\n')

# 2. ì—°ë„ ìŠ¬ë¼ì´ì‹±
df['ì—°ë„'] = df['ë‚ ì§œ'].str[:4]
print(df.head(), '\n')

# 3. ì—°ë„ë³„ ì´ ë²”ì£„ ê±´ìˆ˜ í•© ê³„ì‚° í›„ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
result = df['ì´ë²”ì£„ê±´ìˆ˜'].groupby(df['ì—°ë„']).sum()
result = result.sort_values(ascending=False)
print(result, '\n')     # 2023 ë…„ë„ì˜ ì´ ê±´ìˆ˜ê°€ ê°€ì¥ ë§ì€ ê²ƒ í™•ì¸

# 4. ê°€ì¥ í° ê°’(ì²« ë²ˆì§¸ í–‰)ì˜ ì›” í‰ê·  ê³„ì‚°
print(round(result.iloc[0]/12))
```
