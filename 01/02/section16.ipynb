## section16 그룹핑

```python
import pandas as pd
df = pd.read_csv('./data/cafe4.csv')
df.head(2)
```

### 01. 그룹 기준 1개

```python
df.groupby(['원산지']).mean(numeric_only = True)
```

### 03. 그룹 기준 2개 이상

```python
df.groupby(['원산지', '칼로리']).mean(numeric_only=True)
```

### 04. agg()

```python
df.groupby(['원산지', '메뉴']).agg(['mean', 'sum'])
```

```python
df.groupby(['원산지', '메뉴']).mean(numeric_only = True).reset_index()
```
