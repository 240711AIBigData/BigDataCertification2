```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')
df
```

### 풀이1

```python
# 결측치가 가장 많은 컬럼 확인
df.isnull().sum()
```

```python
# 첫 번째로 결측치가 많은 컬럼에서 결측치가 있는 데이터 제거
df = df.dropna(subset=['f1'])
df
```

```python
# 두 번째로 결측치가 많은 컬럼 최빈값 확인
freq = df['f3'].mode()[0]
freq
```

```python
# 최빈값 대체
df['f3'].fillna(freq, inplace=True)
df.isnull().sum()
```

```python
# f3 컬럼이 gold 인 데이터의 수
print(sum(df['f3'] == 'gold'))
```

### 풀이2

```python
# 결측치가 가장 많은 컬럼
col = df.isnull().sum().idxmax()
col
```

```python
# 결측치가 가장 많은 컬럼 제거
df = df.dropna(subset = [col])
df
```

```python
# 두 번째로 결측치가 많은 컬럼 최빈값 확인
freq = df['f3'].mode()[0]
freq
```

```python
# 최빈값 대체
df['f3'] = df['f3'].fillna(freq)
df.isnull().sum()
```

```python
# f3 컬럼이 gold 인 데이터 수
print(len(df[df['f3'] == 'gold']))
```
