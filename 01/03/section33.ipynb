```python
import pandas as pd
df = pd.read_csv('./data/school_data.csv')
df
```

### 풀이

```python
# 이름을 유지하면서 수학, 영어, 국어 형태 변경
melted_df = df.melt(id_vars=['이름'], value_vars=['수학', '영어', '국어'])
melted_df
```

```python
# 가장 작은 값 25개 합
result = melted_df['value'].nsmallest(25).sum()
print(result)
```
