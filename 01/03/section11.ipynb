```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')
df
```

### 풀이

```python
# 결측치가 있는 행 제거
df.dropna(inplace=True)
df
```

```python
# 70% 지점
end = int(len(df) * 0.7)
end
```

```python
# 70% 데이터 슬라이싱
df = df.iloc[:end]
df
```

```python
# 3사분위수, 1사분위수
q3 = df['views'].quantile(.75)
q1 = df['views'].quantile(.25)
print(q3)
print(q1)
```

```python
# 결과 출력
print(int(q3 - q1))
```
