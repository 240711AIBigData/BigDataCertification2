```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')
df
```

### 풀이

```python
# 결측치 채우기
df = df.fillna(method = 'bfill')
df.isnull().sum()
```

```python
# city와 f2 기준 그룹합 계산
df = df.groupby(['city', 'f2']).sum(numeric_only = True).reset_index()
df
```

```python
# 내림차순 정렬
df = df.sort_values('views', ascending = False)
df
```
