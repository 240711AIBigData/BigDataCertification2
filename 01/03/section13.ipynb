```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')
df
```

```python
# 결측치가 있는 데이터 True/False
cond = df['f1'].isnull()
cond
```

```python
# 결측치가 있는 데이터 선택
df = df[cond]
df
```

```python
# age 평균값
result = df['age'].mean()
result
```

```python
# 소수 첫째 자리까지 출력
print(round(result, 1))
```
