```python
import pandas as pd
df = pd.read_csv('./data/school_data.csv')
df
```

### 풀이

```python
# 수학, 영어, 국어 점수 합계
df['total_score'] = df[['수학', '영어', '국어']].sum(axis=1)
df
```

```python
# 합계 점수를 기준으로 상위 10명 선택
top10 = df.nlargest(10, 'total_score')
top10
```

```python
# 선택된 10명의 수학 평균 점수 계산
result = top10['수학'].mean()
print(round(result))
```
