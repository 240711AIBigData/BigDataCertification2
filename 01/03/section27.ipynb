```python
import pandas as pd
df = pd.read_csv('./data/delivery_time.csv')
df
```

### 풀이

```python
# datetime 자료형으로 변경
df['주문시간'] = pd.to_datetime(df['주문시간'])
print(df['주문시간'])
df['실제도착시간'] = pd.to_datetime(df['실제도착시간'])
print(df['실제도착시간'])
```

```python
# 점심시간 주문 선택
df['시간'] = df['주문시간'].dt.hour
print(df['시간'])

cond1 = df['시간'] >= 10
cond2 = df['시간'] < 13
df = df[cond1 & cond2]
df
```

```python
# 속도(km/h) = 거리(km) / 시간(h)
df['배달시간'] = df['실제도착시간'] - df['주문시간']
print(df['배달시간'])

df['배달시간'] = df['배달시간'].dt.total_seconds() / 60 / 60    # 시간 단위로 변경
print(df['배달시간'])

df['속도'] = df['거리'] / df['배달시간']
sum(df['속도'] >= 50)
```
