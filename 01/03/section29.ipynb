```python
import pandas as pd
df = pd.read_csv('./data/delivery_time.csv')
df
```

### 풀이

```python
# 배달료 계산 함수 정의
def delivery_fee(distance):
    if distance < 5 :
        return 2000
    elif distance < 10 :
        return 4000
    elif distance < 15 :
        return 6000
    elif distance < 20 :
        return 8000
```

```python
# 각 주문에 대한 배달료 계산
df['배달료'] = df['거리'].apply(delivery_fee)
df
```

```python
# 월별로 배달료 총집합 집계
df['주문시간'] = pd.to_datetime(df['주문시간'])
period_M = df['주문시간'].dt.to_period('M')
monthly = df.groupby(period_M)['배달료'].sum()
monthly
```

```python
# 가장 많은 배달료가 발생한 월
max_fee_month = monthly.idxmax()
max_fee_month
```

```python
# 그 월의 총 배달료 찾기
max_fee_value = monthly[max_fee_month]
max_fee_value
```
