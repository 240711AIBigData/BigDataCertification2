```python
import pandas as pd
df = pd.read_csv('./data/type1_data2.csv', index_col = 'year')
df
```

### 풀이1

```python
# 2001, 평균보다 큰 값 합계
m1 = df.loc[2001].mean()
print(m1)

cond1 = df.loc[2001] > m1
print(cond1)

r1 = sum(cond1)
print(r1)
```

```python
# 2003, 평균보다 작은 값 합계
m2 = df.loc[2003].mean()
print(m2)

cond2 = df.loc[2003] < m2
print(cond2)

r2 = sum(cond2)
print(r2)
```

```python
# 두 개수의 합
print(r1 + r2)
```

### 풀이2

```python
# 행과 열 변경
df = df.T
df
```

```python
# 평균보다 큰 값 합계
m1 = df[2001].mean()
cond1 = df[2001] > m1
r1 = sum(cond1)
print(r1)
```

```python
# 평균보다 작은 값 합계
m2 = df[2003].mean()
cond2 = df[2003] < m2
r2 = sum(cond2)
print(r2)
```

```python
# 두 값의 합
print(r1 + r2)
```
