```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')
df
```

### 풀이

```python
# 자료형이 object 컬림 제외
cols = df.select_dtypes(exclude = 'object').columns
df = df[cols]
df
```

```python
# 결측치 0 대체
df = df.fillna(0)
df.isnull().sum()
```

```python
# 행과 열 변경
df = df.T
df
```

```python
# 행별 합이 3000 이상인 데이터 수
print(sum(df.sum() >= 3000))
```
