```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')
df
```

### 풀이

```python
# 자료형 변환
df['subscribed'] = pd.to_datetime(df['subscribed'])
df['subscribed']
```

```python
# 파생 변수 생성(연, 월)
df['year'] = df['subscribed'].dt.year
df['month'] = df['subscribed'].dt.month
df
```

```python
# 2024년 2월이고, 'f3'이 gold 인 데이터
cond1 = df['year']  == 2024
cond2 = df['month']  == 2
cond3 = df['f3'] == 'gold'
df = df[cond1 & cond2 & cond3]
df
```
