```python
import pandas as pd
df = pd.read_csv('./data/type1_data1.csv')
df
```

### 풀이1

```python
# 결측치가 있는 행 삭제
df.dropna(inplace=True)
df
```

```python
# 지역별 평균 계산
df = df.groupby(['city']).mean(numeric_only=True)    # 시험 환경은 mean() 사용
df
```

```python
# f2 컬럼이 가장 큰 지역 출력
print(df.sort_values('f2', ascending=False).index[0])
```
